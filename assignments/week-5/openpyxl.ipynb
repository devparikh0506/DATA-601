```python
import openpyxl
import pandas as pd
```

```python
wb = openpyxl.load_workbook('week_05_homework_XLSX_openpyxl.xlsx')
# wb.save("my_sheet.xlsx")
```

```python
wb.sheetnames
```

```python
pi_sheet = wb['patient info']
```

```python
records_sheet = wb['records']
```

```python
from itertools import islice
def convert_to_df(ws):
    data = ws.values
    cols = next(data)
    data = list(data)
    df = pd.DataFrame(data, columns=cols)
    return df
    
```

```python
pi_df = convert_to_df(pi_sheet)
records_df = convert_to_df(records_sheet)
```

```python
pi_df = pi_df.rename(columns={"patient id":"p_id"})
```

```python
merged_df = pi_df.merge(records_df, how="outer", on="p_id")
merged_df
```

```python
from openpyxl.utils.dataframe import dataframe_to_rows
# pi_sheet.delete_rows(1, pi_sheet.max_row)
merged_df=merged_df.rename(columns={"p_id" : "patient id"})
for row in dataframe_to_rows(merged_df, index=False, header=True) :
    pi_sheet.append(row)
wb.save("my_sheet.xlsx")
```
