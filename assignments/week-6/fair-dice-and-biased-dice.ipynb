```python
import random
import pandas as pd
import plotly.express as px
```

### Creating a function to generate function to simulate fair and biased dice 

```python
def rolldice(num_attempts):
    dice_values=[1,2,3,4,5,6]
    fair=random.choices(dice_values, k=num_attempts)
    biased=random.choices(dice_values, weights=[0.15,0.15,0.15,0.15,0.15,0.25], k=num_attempts)
    return {"fair":fair, "biased":biased }
```

#### Let's create a dataframe with custom input for number of rolls. 

```python
df = pd.DataFrame(rolldice(int(input("Enter number of rolls"))))
```

#### Let's plot histograms for fair and biased columns

```python


fig1 = px.histogram(df['fair'], barmode="group")
fig2 = px.histogram(df['biased'], barmode="group")
fig1.update_layout(bargroupgap=0.1, title="Fair Dice")
fig2.update_layout(bargroupgap=0.1, title="Biased Dice")

fig1.show()
fig2.show()
```

#### From above histograms we can see that for fair dice the counts are alomost same for each number, however for biased dice the counts are almost same for all numbers except 6 which is having a higher bar.

#### Let's polt pie chart to check the percentages of results for each number.

```python
fair_count = df['fair'].value_counts()
pie1 = px.pie(values=fair_count, names=fair_count.index)

biased_count = df['biased'].value_counts()
pie2 = px.pie(values=biased_count, names=biased_count.index)

pie1.update_traces(textposition='inside', textinfo='percent+label', title="Fair Dice")
pie2.update_traces(textposition='inside', textinfo='percent+label', title="Biased Dice")

pie1.show()
pie2.show()
```

##### From above pie charts we can also check that for fair dice the value occurance rate is same for each number and for biased dice the rate is biased towards number '6'.
