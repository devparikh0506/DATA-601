# COUNT LINKS FROM HTML IN XML

### I will use pandas to read xml file convert it into dataframe

```python
import pandas as pd
```

#### Load data from xml file using `pandas.read_xml`

For this assignment I used extracted `xml_containing_html.xml` file.

```python
file_name = "xml_containing_html.xml" # change this file_name to "xml_containing_html.xml.zip" while testing with zip file.
df = pd.read_xml(file_name)
df.head()
```

#### Now importing bs4 library to parse html

```python
from bs4 import BeautifulSoup
```

#### Creating a function that parses html using BeautifulSoup4
I am going to apply this function to `content` column for each row.

**the logic this function uses:**
-  load html using beautifulsoup4 library
- find `<a>` tags using `find_all` method
- return a `pandas.Series` including links count and href values. (This series will be used for creating two new columns)

if this function catches an error then it will return `count=0` and `href=''`

```python
def extract_links_from_html(html):
  try:
    soup = BeautifulSoup(html, 'html.parser')
    links = [link.get("href") for link in soup.find_all('a')]
    return pd.Series([len(links),', '.join(list(set(links)))])
  except Exception as e:
    return pd.Series([0, ''])
```

## Now populating `links_count` and `detected_links` columns to dataframe

`links_count` : total links included in html text

`detected_links` : value of `href` attributes of the detected links

```python
df[["links_count", "detected_links"]] = df["content"].apply(extract_links_from_html)
pd.set_option("display.max_rows", 200)
df
```
