```python
# import xmltodict as xtd
import pandas as pd
```

#### Creating a function to load data from xml file

```python
def get_pages_from_xml(file_name : str):
  try:
    # with open(file_name) as file:
      # data = xtd.parse(file.read())
      # all_pages = data.get("all_pages")
    return pd.read_xml(file_name)
  except:
    return None
```

```python
df = get_pages_from_xml("xml_containing_html.xml")
```

#### Now importing bs4 library to parse html

```python
from bs4 import BeautifulSoup
```

#### Creating a function that parses html using BeautifulSoup4

```python
def extract_links_from_html(html):
  try:
    soup = BeautifulSoup(html, 'html.parser')
    links = [link.get("href") for link in soup.find_all('a')]
    return pd.Series([len(links),', '.join(links)])
  except Exception as e:
    print(e)
    return None
```

```python
df[["links_count", "detected_links"]] = df["content"].apply(extract_links_from_html)
df
```
